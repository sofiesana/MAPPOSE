```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean_returns = np.load("results/_mean_returns.npy")
mean_actor_losses = np.load("results/_mean_actor_losses.npy")
mean_critic_losses = np.load("results/_mean_critic_losses.npy")

# find first index where mean_actor_losses is zero
zero_index = np.where(mean_actor_losses == 0)[0]

# truncate arrays to zero_index 
if len(zero_index) > 0:
    first_zero_index = zero_index[0]
    mean_returns = mean_returns[:first_zero_index]
    mean_actor_losses = mean_actor_losses[:first_zero_index]
    mean_critic_losses = mean_critic_losses[:first_zero_index]

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(mean_returns)
plt.title("Mean Returns over Iterations")
plt.subplot(1,3,2)
plt.plot(mean_actor_losses)
plt.title("Mean Actor Losses over Iterations")
plt.subplot(1,3,3)
plt.plot(mean_critic_losses)
plt.title("Mean Critic Losses over Iterations")
plt.show()
```
